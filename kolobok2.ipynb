```python
def ask_question(question, valid_answers):
    while True:
        answer = input(question + " ").lower()
        if answer in valid_answers:
            return answer
        print("Кажется, вы совсем колобок… Пусть ваша голова полностью состоит из теста, но подумайте внимательно и выберите другой вариант.")

print("КОЛОБОК: Битва за выживание")
print("Тёплое солнечное утро, которое вы встречаете в гостях у бабушки и дедушки, к которым приехали на выходные после сессии. На удивление, вы просыпаетесь не в своей уютной кроватке, человеком, а колобком на подоконнике в избушке. Что же вы собираетесь предпринять?")

while True:
    choice = ask_question("Остаться на подоконнике или укатиться? (остаться/укатиться)", ["остаться", "укатиться"])

    if choice == "остаться":
        print("Вас съели! Игра окончена.")
        break
    else:
        print("Игра начинается! Вы решили укатиться с подоконника.")

    print("\nЧтобы спуститься с подоконника и не превратиться в лепешку, вам необходима помощь. Мимо вашего окна проходит добрый и отзывчивый деревянный мальчишка с азбукой в руках. Позовите его по имени, и он обязательно вам поможет.")
    while True:
        name = input("Как зовут мальчика? ").strip().lower()
        if name == "буратино":
            print("Буратино откликнулся на ваш зов. Спустив вас с подоконника, он поторопился в школу. Вам тоже стоит поторопиться, чтобы скрыться от бабушки с дедушкой.")
            break
        else:
            print("Неправильно! Попробуйте ещё раз.")

    print("\nВы катитесь по лесу и встречаете милого зайчика. Но, кажется, он собирается вас съесть!")
    song = input(
        "Вспомните песню колобка и вставьте пропущенные слова, чтобы вас не съели:\n"
        "“Я колобок, колобок,\n"
        "По амбару метён,\n"
        "По сусечкам скребён,\n"
        "На сметане мешён,\n"
        "В печку сажён,\n"
        "На ... стужён.\n"
        "Я от ... ушёл,\n"
        "Я от ... ушёл,\n"
        "От тебя, зайца, не хитро уйти.”\n"
        "Введите слова по порядку: "
    ).lower().strip()

    if 'окошке' in song and 'дедушки' in song and 'бабушки' in song:
        print("Вы спели песню правильно и укатились от зайца!")
    else:
        print("Вы ошиблись в песне, и заяц вас съел. Игра окончена...")
        break

    print("\nНо вот незадача! На вашем пути встал медведь. Он ищет теремок в лесу. Ваша жизнь снова под угрозой. Чтобы не стать жертвой медведя, помогите ему добраться до теремка, воспользовавшись картой.")
    while True:
        map_choice = input("Введите название карты: гугл, яндекс, я карта ").lower().strip()
        if map_choice == "я карта":
            print("Вы выбрали правильную карту и помогли медведю найти теремок!")
            break
        elif map_choice == "гугл" or map_choice == "яндекс":
            print("Вот незадача! В лесу нет интернета...")
        else:
            print("Что это за карта? Попробуйте другую.")
            
    print("\nВы помогли медведю добраться до теремка, но он не знает, стоит ли заходить внутрь.")
    decision = ask_question("Что вы посоветуете медведю? (войти/остаться)", ["войти", "остаться"])

    if decision == "войти":
        print("Медведь зашёл в теремок. Места не хватило, теремок развалился и раздавил вас! Игра окончена.")
        break
    else:
        print("Медведь разозлился! Вам нужно усыпить его, чтобы укатиться.")
        song = input("Спойте песню, чтобы усыпить медведя (подсказка: в песне должны быть слова 'медведь' и 'спи'): ").strip().lower()

        if "медведь" in song and "спи" in song:
            print("Медведь уснул, и вы смогли укатиться. Вы спасены!")
        else:
            print("Медведь не заснул и съел вас. Игра окончена.")
            break

    print("\nВы спокойно катитесь по тропинке и внезапно врезаетесь в волка. Волк съел бабушку и Красную Шапочку, и теперь за ним гонятся охотники. \nСерый угрожает, что съест и вас, если вы ему не поможете достать бабушку с внучкой из живота. Вы помогаете ему из страха.")
    action = ask_question(
        "Что вы выберете: убежать с волком или дождаться охотников и получить волшебный приз? (убежать/дождаться)",
        ["убежать", "дождаться"]
    )

    if action == "убежать":
        print("Вы убежали с волком, но он проголодался и съел вас. Игра окончена.")
        break
    else:
        print("Вы дождались охотников, и они помогли бабушке и Красной Шапочке. Волшебный мешок теперь у вас!")

    print("\nВы продолжаете свой путь и встречаете хитрую лису. Она начинает рассказывать, какой вы молодец, как хорошо вы закроете сессию без усилий. \nЗатем она спрашивает, что вы здесь делаете один.")
    choice = ask_question(
        "Что вы хотите сделать? (подойти и ответить/поискать в мешочке волшебный предмет): ",
        ["подойти и ответить", "поискать в мешочке волшебный предмет"]
    )

    if choice == "подойти и ответить":
        print("\nЛиса заговорила вам зубы и проглотила. Игра окончена.")
        break
    else:
        print("\nВы молча слушаете и незаметно тянетесь в волшебный мешок. Там лежат три предмета:")
        print("Меч-кладенец\nДудочка крысолова\nСтранное Зелье")
        while True:
            choice_item = input("Выберите предмет (меч/дудочка/зелье): ").lower()
            if choice_item == "меч":
                print("Вы сразили лису одним ударом меча. Она больше не угрожает вам!")
                break
            elif choice_item == "дудочка":
                print("Вы заиграли на дудочке, и лиса уснула. Вы спокойно укатились.")
                break
            elif choice_item == "зелье":
                print("Вы выпили Странное Зелье, стали огромным качком-колобком и победили лису врукопашную!")
                break
            else:
                print("Неверный выбор. Попробуйте ещё раз.")
    print("Вы победили лису и можете продолжить своё путешествие.")
    print("\nУставший и грустный, вы встречаете на своём пути just a chill guy. Chill guy хочет проверить, насколько вы chill. \nОтветьте на пару его вопросов:")
    questions = {
        "Кто первым совершил кругосветное путешествие?": "Фернан Магеллан",
        "Назовите год основания ВШЭ": "1992",
        "Первое правило бойцовского клуба?": "Никому не рассказывать о бойцовском клубе",
        "Дополни фразу: Без ... не выловишь и рыбку из пруда": "труда",
        "Сколько будет 5*5": "25"
    }
    ball = 0
    for question, answer in questions.items():
        print(question)
        if input("Ваш ответ: ").strip() != answer:
            ball += 1

    if ball <= 1:
        print("Дружище, надо расслабиться!")
    elif ball <= 3:
        print("Basic уровень…")
    else:
        print("Ученик превзошел учителя!")

    print("\nНемного почиллив, вы продолжаете катиться и катиться по лесу и упираетесь в поросёнка. \nОн очень напряжён, ведь ему предстоит заняться строительством. Помогите поросёнку построить домик, чтобы он не съел вас.")
    material = ask_question(
        "Из какого материала будем строить дом? (ветки/кирпичи/сено/пряники)",
        ["ветки", "кирпичи", "сено", "пряники"]
    )
    
    if material == "кирпичи":
        print("Вы построили кирпичный дом. Волк не сможет его разрушить. Вы в безопасности!")
    elif material in ["ветки", "сено"]:
        print("Вы построили дом из веток или сена. Волк сдул его и съел вас. Конец игры.")
        break
    elif material == "пряники":
        print("Вы построили пряничный дом. Волк съел ваш дом, наелся пряниками и ушёл, не тронув вас.")
        print("Поросёнок расстроен, что у него больше нет дома. Но вы все живы, ура ура!")
        print("Чтобы утешить поросёнка, вы поёте песню и катитесь дальше.")
    
    print("\nВаш путь был таким долгим, что успел пойти снег… \nИ вот Вы встречаете Деда Мороза, он расстроен, потому что потерял письма от обитателей леса и теперь не знает, что им дарить. \nСреди лесных обитателей: лисенок, зайчонок, медвежонок, белочка, дедушка и волк. \nВы можете помочь Деду Морозу, указав, какие подарки из мешочка дедушки нужно подарить этим зверятам. \nВ мешке у деда мороза есть: айфон 16, мёд, норковая шуба, капуста, дайсон, арахис, золотые орешки с изумрудными ядрами, черника, репка. ")
    selection = ask_question(
        "Что вы выберете: помочь или пройти мимо? (помочь/пройти мимо)",
        ["помочь", "пройти мимо"]
    )
    if selection == "пройти мимо":
        print("Из-за своей черствости вы навсегда останетесь засохшей буханкой. Игра окончена.")
        break
    else:
        print("\nВ списке Деда Мороза оказался кто-то лишний. Он собирается подарить подарки лисенку, зайчонку, медвежонку, белочке, дедушке и волку. ")
    animals = ["зайчику", "лисёнку", "медвежонку", "белочке", "дедушке", "волку"]
    gifts = {
        "зайчику": "капусту",
        "лисёнку": "чернику",
        "медвежонку": "мёд",
        "белочке": "золотые орешки с изумрудными ядрами",
        "дедушке": "репку",
    }
    a = input("Кому не стоит ничего дарить из-за плохого поведения?")
    if a == 'волку':
        print("Правильно! Туда этого вшивого.")
        animals.remove("волку")
    else:
        print("Неверно. Немного подумав, Дед Мороз решил не дарить подарок волку. Туда этого вшивого.")
        animals.remove("волку")
    while True:
        new_gifts = {}
        for animal in animals:
            gift = input(f"Что подарить {animal}? ")
            new_gifts[animal] = gift
        if new_gifts == gifts:
            print("Вы успешно помогли Деду Морозу!")
            break
        else:
            print("Подарки не подошли, попробуйте снова.")
    print("\nНаконец, вы добрались до золотой рыбки, живущей в волшебном озере в лесу.")
    wish = ask_question(
        "Вы можете загадать желание: стать человеком, круассаном или рафом",
        ["человеком", "круассаном", "рафом"]
    )

    if wish == "человеком":
        print("Поздравляю! Вы дошли до конца и вернули себе человеческий облик! Возьмите с полки пирожок, если еще не воротит от мучного!")
    elif wish == "круассаном":
        print("Вас продадут в Джеффрис на съедение преподавателю по Питону.")
    else:
        print("Вас выпьет студент в Груше!")
    break
```
